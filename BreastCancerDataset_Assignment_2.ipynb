```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

Data preperation

```python
dataset.describe()
```

```python
dataset.isnull().sum() # Checking if null values are present
```

```python
dataset.shape   #Rows and Attributes
```

```python
from sklearn.model_selection import train_test_split
X = dataset.copy()
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=4)
clf = clf.fit(X_train, y_train)
```

```python
clf.get_params()
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
clf.predict_proba(X_test)
```

Feature scaling

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, labels=[0,1])
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, predictions)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names=['malignant', 'benign'] ))
```

```python
feature_names = X.columns
feature_names
```

```python
from sklearn import tree
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf,
                   feature_names=feature_names,
                   class_names={0:'Malingnant', 1:'Benign'},
                   filled=True,
                   fontsize=12)
```

Optimizing Decision Tree Using Entropy

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=1,ccp_alpha=0.01)
clf = clf.fit(X_train, y_train)

from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf,
                   feature_names=feature_names,
                   class_names={0:'Malingnant', 1:'Benign'},
                   filled=True,
                   fontsize=12)
```
